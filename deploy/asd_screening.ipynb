```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# -----------------------------------------
# 1) Use age only (df1 = behavioral, df2 = clinical)
# -----------------------------------------
age_beh = df1[['Age_Years']].astype(float).values
age_clin = df2[['Age_Years']].astype(float).values

# Scale based on df2 (donors)
scaler = StandardScaler().fit(age_clin)
age_beh_s = scaler.transform(age_beh)
age_clin_s = scaler.transform(age_clin)

# -----------------------------------------
# 2) Nearest neighbor matcher (1 clinical donor per behavioral child)
# -----------------------------------------
nn = NearestNeighbors(n_neighbors=1).fit(age_clin_s)
dist, idx = nn.kneighbors(age_beh_s)

# -----------------------------------------
# 3) Select clinical columns to attach
# -----------------------------------------
exclude = {'Age_Years','SITE_ID','SUB_ID','NDAR_GUID'}
clin_cols = [c for c in df2.columns if c not in exclude]

attached = df2.iloc[idx.flatten()][clin_cols].reset_index(drop=True)
attached = attached.add_suffix('_clin')

# -----------------------------------------
# 4) Final merged dataset (df)
# -----------------------------------------
df = pd.concat([df1.reset_index(drop=True),
                attached.reset_index(drop=True)], axis=1)

df['donor_nn_distance'] = dist.flatten()
```

```python
from sklearn.experimental import enable_iterative_imputer   # MUST COME FIRST
from sklearn.impute import SimpleImputer, IterativeImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, RobustScaler
from sklearn.model_selection import StratifiedKFold, GridSearchCV, cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, confusion_matrix, ConfusionMatrixDisplay)
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")



# --------------- 1) Prepare X, y, and column lists (no new dfs) ---------------
target = "DX_GROUP_clin"
X = df.drop(columns=[target])
y = df[target].astype(int)

# numeric / categorical lists (computed from current df)
numeric_columns = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_columns = X.select_dtypes(include=['object','category']).columns.tolist()

# split categorical columns into binary (<=2 unique vals) vs multi (>2)
binary_cats = [c for c in categorical_columns if X[c].nunique(dropna=True) <= 2]
multi_cats  = [c for c in categorical_columns if X[c].nunique(dropna=True) > 2]

print(f"{len(numeric_columns)} numeric cols, {len(binary_cats)} binary categorical, {len(multi_cats)} multi-category")

# --------------- 2) Build preprocessing ColumnTransformer ---------------
# numeric pipeline: IterativeImputer (multivariate) -> RobustScaler
num_pipeline = Pipeline([
    ("imp", IterativeImputer(random_state=0, max_iter=10)),
    ("scale", RobustScaler())
])

# binary categorical pipeline: SimpleImputer (most frequent) -> OrdinalEncoder
bin_pipeline = Pipeline([
    ("imp", SimpleImputer(strategy="most_frequent")),
    ("ord", OrdinalEncoder())
])

# multi categorical pipeline: SimpleImputer -> OneHotEncoder(handle_unknown='ignore')
multi_pipeline = Pipeline([
    ("imp", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer([
    ("num", num_pipeline, numeric_columns),
    ("bin_cat", bin_pipeline, binary_cats),
    ("multi_cat", multi_pipeline, multi_cats)
], remainder="drop", sparse_threshold=0)

# --------------- 3) Build full pipeline with selection + placeholder classifier ---------------
selector = SelectFromModel(RandomForestClassifier(n_estimators=200, random_state=0), threshold="median")

base_pipeline = Pipeline([
    ("preproc", preprocessor),
    ("select", selector),
    ("clf", RandomForestClassifier(random_state=0))   # placeholder; GridSearch will replace
])

# --------------- 4) Setup CV and model grids ---------------
cv_inner = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
cv_outer = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)  # for OOF predictions

grids = {}

# Logistic Regression grid (use 'liblinear' for small/medium data)
grids['logreg'] = {
    'clf': [LogisticRegression(solver='liblinear', max_iter=200, random_state=0)],
    'clf__C': [0.01, 0.1, 1.0, 10.0],
    'clf__penalty': ['l2']
}

# Random Forest grid
grids['rf'] = {
    'clf': [RandomForestClassifier(random_state=0, n_jobs=-1)],
    'clf__n_estimators': [100, 300],
    'clf__max_depth': [None, 10, 20],
    'clf__class_weight': [None, 'balanced']
}

# Gradient Boosting grid
grids['gb'] = {
    'clf': [GradientBoostingClassifier(random_state=0)],
    'clf__n_estimators': [100, 200],
    'clf__learning_rate': [0.01, 0.1],
    'clf__max_depth': [3, 5]
}

# --------------- 5) Run GridSearchCV for each model separately ---------------
best_results = {}
for name, param_grid in grids.items():
    print(f"\nRunning GridSearchCV for {name} ...")
    gs = GridSearchCV(base_pipeline, param_grid, scoring='roc_auc', cv=cv_inner,
                      n_jobs=-1, verbose=0, refit=True)
    gs.fit(X, y)
    best_results[name] = gs
    print(" Best ROC-AUC (mean,cv) =", round(gs.best_score_,4))
    print(" Best params:", gs.best_params_)

# --------------- 6) Choose best model by best_score_ (roc_auc) ---------------
best_name = max(best_results.items(), key=lambda kv: kv[1].best_score_)[0]
best_gs = best_results[best_name]
print(f"\nBEST MODEL: {best_name} with ROC-AUC = {best_gs.best_score_:.4f}")

# --------------- 7) Get out-of-fold predictions (no leakage) using cross_val_predict ---------------
# cross_val_predict will clone the estimator and refit inside each fold (good).
best_estimator = best_gs.best_estimator_

y_pred = cross_val_predict(best_estimator, X, y, cv=cv_outer, method='predict', n_jobs=-1)
y_prob = cross_val_predict(best_estimator, X, y, cv=cv_outer, method='predict_proba', n_jobs=-1)[:, 1]

# --------------- 8) Metrics ---------------
acc = accuracy_score(y, y_pred)
prec = precision_score(y, y_pred, zero_division=0)
rec = recall_score(y, y_pred, zero_division=0)
f1 = f1_score(y, y_pred, zero_division=0)
roc = roc_auc_score(y, y_prob)

print("\n=== Final OOF metrics (Stratified 5-fold) ===")
print(f"Accuracy : {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall   : {rec:.4f}")
print(f"F1-score : {f1:.4f}")
print(f"ROC-AUC  : {roc:.4f}")

# --------------- 9) Confusion matrix plot ---------------
cm = confusion_matrix(y, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_estimator.named_steps['clf'].classes_)
fig, ax = plt.subplots(figsize=(5,5))
disp.plot(ax=ax, cmap='Blues', colorbar=False)
plt.title(f"Confusion Matrix — {best_name}")
plt.show()

# --------------- 10) Retrain best pipeline on FULL data (final model object saved) ---------------
final_model = best_gs.best_estimator_
final_model.fit(X, y)   # fits preprocessor, selector, and classifier on all data

print("\nFinal model refit on full data. Use `final_model` for inference (predict / predict_proba).")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix
)
import matplotlib.pyplot as plt

# ----------------------------
# 1. Build subgroups
# ----------------------------

df_fair = df.copy()
df_fair["y_true"] = y
df_fair["y_pred"] = y_pred
df_fair["y_prob"] = y_prob

# AGE GROUPS
def age_group(x):
    if x < 12:
        return "Child"
    elif x <= 17:
        return "Teen"
    else:
        return "Adult"

df_fair["AGE_GROUP"] = df_fair["Age_Years"].apply(age_group)

# SEX GROUPS
df_fair["SEX_GROUP"] = df_fair["Sex"].fillna("Unknown")

# ETHNICITY GROUP
df_fair["ETH_GROUP"] = df_fair["Ethnicity"].fillna("Unknown")


# ----------------------------
# 2. Function to compute metrics
# ----------------------------

def subgroup_metrics(df, group_col):
    results = []
    for group in df[group_col].unique():
        sub = df[df[group_col] == group]
        if len(sub) < 5:
            continue   # skip tiny groups (unstable)

        acc = accuracy_score(sub.y_true, sub.y_pred)
        prec = precision_score(sub.y_true, sub.y_pred, zero_division=0)
        rec = recall_score(sub.y_true, sub.y_pred, zero_division=0)   # SENSITIVITY
        f1 = f1_score(sub.y_true, sub.y_pred, zero_division=0)

        try:
            auc = roc_auc_score(sub.y_true, sub.y_prob)
        except:
            auc = np.nan

        cm = confusion_matrix(sub.y_true, sub.y_pred)
        if cm.shape == (2, 2):
            tn, fp, fn, tp = cm.ravel()
            fpr = fp / (fp + tn + 1e-9)
            fnr = fn / (fn + tp + 1e-9)
        else:
            # Only one class present in y_true/y_pred
            tn = fp = fn = tp = np.nan
            fpr = fnr = np.nan

        results.append([group, len(sub), acc, prec, rec, f1, auc, fpr, fnr])

    return pd.DataFrame(results, columns=[
        group_col, "N", "Accuracy", "Precision", "Recall(Sensitivity)",
        "F1", "ROC-AUC", "FPR", "FNR"
    ])


# ----------------------------
# 3. Compute fairness tables
# ----------------------------

sex_table = subgroup_metrics(df_fair, "SEX_GROUP")
age_table = subgroup_metrics(df_fair, "AGE_GROUP")
eth_table = subgroup_metrics(df_fair, "ETH_GROUP")

print("\n=== SEX FAIRNESS ===")
print(sex_table)

print("\n=== AGE FAIRNESS ===")
print(age_table)

print("\n=== ETHNICITY FAIRNESS ===")
print(eth_table)


# ----------------------------
# 4. Bar plots — recall (sensitivity)
# ----------------------------

def plot_metric(table, group_col, metric, title):
    plt.figure(figsize=(6,4))
    plt.bar(table[group_col], table[metric], color="skyblue")
    plt.ylabel(metric)
    plt.title(title)
    plt.xticks(rotation=45)
    plt.ylim(0,1)
    plt.show()

plot_metric(sex_table, "SEX_GROUP", "Recall(Sensitivity)", "Recall by Sex")
plot_metric(age_table, "AGE_GROUP", "Recall(Sensitivity)", "Recall by Age")
plot_metric(eth_table, "ETH_GROUP", "Recall(Sensitivity)", "Recall by Ethnicity")


# ----------------------------
# 5. Confusion matrix per subgroup
# ----------------------------

def plot_subgroup_cm(df, group, group_name):
    sub = df[df[group] == group_name]
    if len(sub) < 5:
        print(f"Skipping {group_name} (too few samples)")
        return
    cm = confusion_matrix(sub.y_true, sub.y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix — {group}: {group_name}")
    plt.show()

for g in df_fair["SEX_GROUP"].unique():
    plot_subgroup_cm(df_fair, "SEX_GROUP", g)

for g in df_fair["AGE_GROUP"].unique():
    plot_subgroup_cm(df_fair, "AGE_GROUP", g)

for g in df_fair["ETH_GROUP"].unique():
    plot_subgroup_cm(df_fair, "ETH_GROUP", g)
```

```python
# Feature importance extractor + plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

def _get_feature_names_from_column_transformer(ct: ColumnTransformer, input_columns):
    """
    Return a list of output feature names from a ColumnTransformer `ct`.
    `input_columns` should be the X DataFrame columns (pd.Index or list).
    """
    feature_names = []
    for name, transformer, cols in ct.transformers_:
        if name == 'remainder':
            # if remainder='passthrough' include those columns
            if transformer == 'passthrough' or transformer == None:
                # determine remaining columns
                passed = [c for c in input_columns if c not in ct._columns]
                feature_names.extend(passed)
            continue

        # cols can be slice, boolean mask, list etc. Normalize to list of names
        if isinstance(cols, (list, tuple, np.ndarray)):
            cols_names = list(cols)
        else:
            # assume string column
            cols_names = [cols]

        # If transformer is a pipeline, get the final transformer
        transf = transformer
        try:
            # pipeline-like
            if hasattr(transformer, 'steps'):
                transf = transformer.steps[-1][1]
        except Exception:
            pass

        # If transformer exposes get_feature_names_out -> use it (OneHotEncoder, etc.)
        try:
            # many transformers accept input_features to get_feature_names_out
            names = transf.get_feature_names_out(cols_names)
            feature_names.extend(list(names))
            continue
        except Exception:
            pass

        # Fallbacks:
        # - numeric pipelines (no expansion): keep original column names
        feature_names.extend(cols_names)

    return feature_names

# --- main logic ---
# final_model must be a fitted Pipeline: [ ("preproc", ColumnTransformer), ("select", SelectFromModel), ("clf", estimator) ]
if not isinstance(final_model, Pipeline):
    raise RuntimeError("`final_model` should be a fitted sklearn Pipeline (preproc -> select -> clf).")

# try to get preprocessor, selector, classifier
preproc = final_model.named_steps.get('preproc', None)
selector = final_model.named_steps.get('select', None)
clf = final_model.named_steps.get('clf', None)

if preproc is None or not isinstance(preproc, ColumnTransformer):
    raise RuntimeError("Pipeline does not contain a ColumnTransformer named 'preproc' as expected.")

# X must be the dataframe with features used earlier
input_columns = list(X.columns)

# Give ColumnTransformer a helper attribute listing original columns used (best-effort)
# (This makes fallback remainder handling safer if ct._columns not present)
try:
    # store original columns for fallback usage (only used above in function)
    preproc._columns = input_columns
except Exception:
    pass

# 1) get expanded feature names
all_feature_names = _get_feature_names_from_column_transformer(preproc, input_columns)

# 2) apply selector mask if present
if selector is not None:
    try:
        mask = selector.get_support()
        # if mask length mismatches, fallback to all features
        if len(mask) != len(all_feature_names):
            print("Warning: selector support length doesn't match feature name count; skipping selector masking.")
            selected_feature_names = all_feature_names
        else:
            selected_feature_names = [f for f,m in zip(all_feature_names, mask) if m]
    except Exception as e:
        print("Warning: could not apply selector.get_support(). Using all features. (", e, ")")
        selected_feature_names = all_feature_names
else:
    selected_feature_names = all_feature_names

# 3) get importances from classifier
if clf is None:
    raise RuntimeError("Pipeline has no classifier step named 'clf'.")

importances = None
importance_type = None

# Linear model with coef_
if hasattr(clf, "coef_"):
    coef = clf.coef_
    # binary classifier => coef_.shape = (1, n_features) or (n_classes, n_features)
    if coef.ndim == 1 or coef.shape[0] == 1:
        arr = coef.ravel()
    else:
        # multiclass: take absolute mean across classes
        arr = np.mean(coef, axis=0)
    importances = np.abs(arr)
    importance_type = "abs(coef)"
# tree-based model
elif hasattr(clf, "feature_importances_"):
    importances = clf.feature_importances_
    importance_type = "feature_importance"
else:
    raise RuntimeError("Classifier has neither coef_ nor feature_importances_. Can't get importances.")

# 4) map importances -> selected_feature_names
if len(importances) != len(selected_feature_names):
    print(f"Warning: number of importances ({len(importances)}) != number of feature names ({len(selected_feature_names)}).")
    # try to align by trimming or padding zeros
    L = min(len(importances), len(selected_feature_names))
    importances = importances[:L]
    selected_feature_names = selected_feature_names[:L]

feat_imp = pd.Series(importances, index=selected_feature_names)
feat_imp_sorted = feat_imp.sort_values(ascending=False)

# 5) display top features and plot
top_n = 30
print("\nTop features (by {}):\n".format(importance_type))
print(feat_imp_sorted.head(top_n))

# Plot horizontal bar
plt.figure(figsize=(8, min(0.3 * top_n + 2, 18)))
feat_imp_sorted.head(top_n).sort_values().plot(kind='barh')
plt.title(f"Top {top_n} features ({importance_type})")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()
```
